```python
def cdf(sample, x, sort = False):
    # Sorts the sample, if unsorted
    if sort:
        sample.sort()
    # Counts how many observations are below x
    cdf = sum(sample <= x)
    # Divides by the total number of observations
    cdf = cdf / len(sample)
    return cdf
```

```python
def ks_norm(sample):
    # Sorts the sample
    sample.sort()
    # Evaluates the KS statistic
    D_ks = [] # KS Statistic list
    for x in sample:
        cdf_normal = stats.norm.cdf(x = x, loc = 0, scale = 1)
        cdf_sample = cdf(sample = sample, x  = x)
        D_ks.append(abs(cdf_normal - cdf_sample))
    ks_stat = max(D_ks)
    # Calculates the P-Value based on the two-sided test
    # The P-Value comes from the KS Distribution Survival Function (SF = 1-CDF)
    p_value = stats.kstwo.sf(ks_stat, len(sample))
    return {"ks_stat": ks_stat, "p_value" : p_value}
```

```python
def ks_2samp(sample1, sample2):
    # Gets all observations
    observations = np.concatenate((sample1, sample2))
    observations.sort()
    # Sorts the samples
    sample1.sort()
    sample2.sort()
    # Evaluates the KS statistic
    D_ks = [] # KS Statistic list
    for x in observations:
        cdf_sample1 = cdf(sample = sample1, x  = x)
        cdf_sample2 = cdf(sample = sample2, x  = x)
        D_ks.append(abs(cdf_sample1 - cdf_sample2))
    ks_stat = max(D_ks)
    # Calculates the P-Value based on the two-sided test
    # The P-Value comes from the KS Distribution Survival Function (SF = 1-CDF)
    m, n = float(len(sample1)), float(len(sample2))
    en = m * n / (m + n)
    p_value = stats.kstwo.sf(ks_stat, np.round(en))
    return {"ks_stat": ks_stat, "p_value" : p_value}
```

```python
sample_1 = np.array(df[df['base'] == 'hbo']['imdb_score'].dropna().sample(1000))
sample_1.sort()

cdf_1 = [cdf(sample_1, x) for x in sample_1]

sample_2 = np.array(df[df['base'] == 'disney']['imdb_score'].dropna().sample(1000))
sample_2.sort()

cdf_2 = [cdf(sample_2, x) for x in sample_2]

print(ks_2samp(sample_1, sample_2))

plt.step(
    sample_1,
    cdf_1,
    label='hbo distribution'
)


plt.step(
    sample_2,
    cdf_2,
    label='disney distribution'
)
plt.legend()
plt.show()
```

```python
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np

def get_statistic(sample, population_cdf):
    emperical = [(1 + i) / len(sample) for i in range(len(sample))]
    real = [population_cdf(s) for s in sample]
    return max([abs(e-r) for e,r in zip(emperical, real)])

shape = 2
scale = 2
population_cdf = lambda x: gamma.cdf(x, shape, scale=scale)
xs = np.arange(0, 20, 0.01)
plt.plot(
    xs,
    [population_cdf(x) for x in xs],
    label='actual distribution'
)

n = 80
sample = np.sort(np.random.gamma(shape, scale, n))
emperical = [(1 + i) / len(sample) for i in range(len(sample))]

statistic = get_statistic(sample, population_cdf)
print('statistic: ', statistic)
plt.step(
    sample,
    emperical,
    label='empirical distribution'
)
plt.legend()
plt.show()
```
